```python
import matplotlib.pyplot as plt
```

```python
#the objective is to compare the analytical results to the numerical (Euler) results
#initialize variables
g,h = 9.8,0.02 #gravity and time step
y,v_0=0.0,5.0 #initial position and velocity
t_a,y_a = [],[]#an array to store time and position for the numerical calculation
t,yb = 0.0,[] #initial time=0 , an array to store the results of the analytic results

#loop to calculate a numeric and analytic result at a time t
```

```python
while t<1.0:
    t_a.append(t)#record time
    y_a.append(y)#record position(numeric result)
    yb.append(v_0*t-(g*t**2)/2.0)#analytic result
    
    v=v_0-g*t #calculate new velocity
    y=y+v*h #calculate new position
    t=t+h #move time forward by time step
```

```python
plt.figure()
plt.plot(t_a,y_a,t_a,yb,'--')
plt.show()
```
